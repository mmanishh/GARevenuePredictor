```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
from pandas.io.json import json_normalize


pd.set_option('display.max_columns', 100)


% matplotlib inline
```

```python
train = pd.read_csv('data/train.csv',nrows=10000)
```

```python
train.head()
```

```python
columns_toprocess = ['device','geoNetwork','totals','trafficSource']
```

```python
def json2columns(train,columns):
    for column in columns:
        train= train.join(train[column].apply(json.loads).apply(pd.Series))
    return train
```

```python
new_train = json2columns(train,columns_toprocess)
```

```python
new_train.drop(columns=columns_toprocess,inplace=True)
```
